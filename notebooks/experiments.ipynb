```python
from rag.data_loader import load_documents

documents = load_documents("../data")

len(documents)
```

```python
from rag.chunking import chunk_documents

chunks = chunk_documents(
        documents,
        chunk_size=500,
        chunk_overlap=100
    )

len(chunks)
```

```python
chunks[0]
```

```python
from rag.embedding import create_embeddings

embedded_chunks = create_embeddings(chunks)
```

```python
from rag.vector_store import create_faiss_index

index = create_faiss_index(embedded_chunks)
```

```python
from rag.retriever import retrieve_chunks
from rag.prompt_builder import build_prompt
from rag.generator import generate_answer
import textwrap

query = "খলিশা ফুলের মধু price কত?" 

results = retrieve_chunks(
    query= query,
    index=index,
    embedded_chunks=embedded_chunks,
    top_k=3
)
# print(results)

prompt = build_prompt(
    query=query,
    retrieved_results=results
)
# print(prompt)

answer = generate_answer(prompt)

wrapped_answer = textwrap.fill(answer, width=80)

print(wrapped_answer)
```
